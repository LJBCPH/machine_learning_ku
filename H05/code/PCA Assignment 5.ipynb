```python
import numpy as np
from sklearn.decomposition import PCA
from sklearn import datasets

import matplotlib.pyplot as plt
%matplotlib inline
```

Load handwritten digits data:

```python
digits = datasets.load_digits()
imshape = digits.images[0].shape  # The 2D image shape
X = digits.images.reshape(digits.images.shape[0],-1)  # Image
```

Do the PCA:

```python
pca = PCA()
pca.fit(X)
eigenvalues = pca.explained_variance_
explained_variance = eigenvalues / np.sum(eigenvalues)
num_eig_vec = np.arange(1, len(X.T)+1, 1)

# Explained variance in terms of eigenvalues
fig, ax = plt.subplots(figsize = (6, 6))
plt.plot(num_eig_vec, explained_variance)
plt.xlabel("number of eigenvector")
plt.ylabel("eigenvalue")
ax.set_yscale('log')
#plt.semilogy()
fig.savefig('C:/Users/lucas/OneDrive/Skrivebord/repo/machine_learning_ku/H05/plots/h05_explained_variance.png')
plt.show()

# Cummulative explained variance
cum_exp_var = np.cumsum(explained_variance)
fig, ax = plt.subplots(figsize = (6, 6))
plt.plot(num_eig_vec, cum_exp_var)
plt.xlabel("number of eigenvector")
plt.ylabel("eigenvalue")
plt.axvline(x=10)
fig.savefig('C:/Users/lucas/OneDrive/Skrivebord/repo/machine_learning_ku/H05/plots/h05_cummulated_explained_variance.png')
plt.show()
print(np.sum(explained_variance[0:10]))

#print("'Explained variance' (not normalized):\n", pca.explained_variance_)
#print("Explained variance per component (computed from previous result):\n", pca.explained_variance_ / np.sum(pca.explained_variance_))
#print("Explained variance per component:\n", pca.explained_variance_ratio_)
```

```python
# Eigenfaces
components_5 = pca.components_[0:5, :]
def plot_img(img, savepath):
    example_pic = img
    example_pic = np.reshape(example_pic, (8, 8))
    fig = plt.figure()
    plt.imshow(example_pic, cmap = 'gray_r')
    plt.colorbar()
    plt.grid(True)
    fig.savefig(savepath, bbox_inches='tight')
print(components_5[0,:])
for i in range(5):
    plot_img(img = components_5[i, :], savepath = 'C:/Users/lucas/OneDrive/Skrivebord/repo/machine_learning_ku/H05/plots/digits_img_plot_{0}.png'.format(i))
#plt.imshow(components_5[1,:].reshape(imshape), interpolation='nearest')
```
